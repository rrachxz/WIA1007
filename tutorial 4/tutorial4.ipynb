```python
import matplotlib.pyplot as plt
```

```python
top_artists = data.sort_values(by="artist_followers", ascending=False).head(5)

popularity = top_artists["artist_popularity"]
followers = top_artists["artist_followers"]

plt.bar(popularity, followers)
plt.xlabel(popularity) 
plt.ylabel (followers)
plt.show()
```
