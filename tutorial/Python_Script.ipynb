Containers

List, Dictionaries, Set, Tuples

### List

```python
list1=[3,12,11, 14, 9]
print(list1, list1[2])
```

```python
print(list1[-1])
print(list1[4])#writing at index 2 or fetching data from index 2
print(list1)
```

```python
list1[3]='data' # replacing the value present at index 3
print(list1)
```

```python
list1.append('Science')
print(list1)
#the size of list is flexible
```

```python
list1.pop(2)# deletes the last element
print(list1)
```

```python
numbers=list(range(6))
print(numbers)
```

```python
#slicing
list2=[11, 2, 5, 18, 3, 20]
print(list2[2:6]) # value at index 2 and 3
print(list2[2:])
print(list2[:4])
print(list2[:])
print(list2[:-2])
list2[2:4]= [50,100]
print(list2)
```

### Dictionary

```python
d={'cat':'cute', 'dog':'furry'}#key-value, key can be of any type
print(d['cat'])
print('cat' in d)
```

```python
d['dog']='not furry'
print(d)
```

```python
del d['cat']
print(d)
```
